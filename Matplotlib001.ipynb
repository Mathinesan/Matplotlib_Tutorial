```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,5,7,9,11,19,25,34,45,55]
y = [12,24,36,48,52,64,69,72,86,91]
plt.plot(x, y)
plt.show()
```

```python
print(plt.style.available)
#plt.savefig('image.png')
```

```python
plt.plot(x, y , linestyle='dotted')
plt.show()
```

```python
plt.plot(x, y , linestyle='dotted')
plt.title("no. of vehicle travel in path")
plt.xlabel("Total number of path")
plt.ylabel("Number of vehicle")
plt.show()
```

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.bar(x, y)
plt.show()
```

```python
plt.hist(x)
plt.hist(y)
plt.show()
```

```python
lable = ["Apples", "Bananas", "Cherries", "Dates"]
explodes = [0, 0.5, 0, 0]
plt.pie(y, labels=lable, explode = explodes)
plt.show()
```
